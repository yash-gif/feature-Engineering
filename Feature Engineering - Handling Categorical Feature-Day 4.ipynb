### Ordinal measurement advantages
ordinal measurement is normally used for surveys and questionnaires. Statistical analysis is applied to the responses once they are collected to place the people who took the survey into the various categories. The data is then compared to draw inferences and conclusions about the whole surveyed population with regard to the specific variables. The advantages of using ordinal measurement is ease of collation and categorization.If you ask a seuvey question without providing the variables, the answers are likely to be so diverse they cannot be converted to statistics.

#### with respect to machine learning

1. keep the semantical information of the variable (human readable content)
2.Straightforward

### Ordinal Measurement Disadvantages
The same characteristics of ordinal measurement that create its advantages also create certain disadvantages . THe responses are often so narrow in relation to the question that they create or magnify blas that is not factored into the survey. For example, on the question about satisfaction with the governor , people might be satisfied with his job performance but upset about a recent sex scandal. The servey question might lead respondents to state their dissatisfaction about the scandal, inspite of satisfaction with his job performance-- but the statistical conclusion will not differentiate.

#### with respect to machine learning
1. Does not add machine learning valuable information.

#### Count Or Frequency Encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header= None,index_col=None) 
train_set.head()
```

```python
train_set[1].unique()
```

```python
len(train_set[1].unique())
```

```python
columns=[1,3,5,6,7,8,9,13]
```

```python
train_set=train_set[columns]
```

```python
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()),'labels')
```

```python
train_set['Country'].value_counts()
```

```python
country_map=train_set['Country'].value_counts().to_dict()
country_map
```

```python
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

#### Target Guided Ordinal Encoding
1.Ordering the labels according to the target

2.Replace the labels by the joint probability of being 1 or 0

```python
df=pd.read_csv('Titanic_dataset.csv',usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df.head(20)
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_label2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_label2
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_label2)
```

```python
df.head()
```

```python
df.drop(['Cabin'],axis=1,inplace=True)
```

```python
df.head()
```

#### Mean Encoding

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
```
