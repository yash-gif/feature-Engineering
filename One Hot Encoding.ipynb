```python
import numpy as np
import pandas as pd

data = pd.read_csv("C:\\Users\\lenovo\\Downloads\\train.csv.zip" , usecols=['X1','X2','X3','X4','X5','X6'])
data.head()
```

```python
# one Hot encoding - variables with many variables
```

```python
# let look how many labels each variables has:

for col in data.columns:
    print(col, ":" , len(data[col].unique()),"labels")
```

```python
# lets examine how many columns we will obtain after one hot encoding these variables

pd.get_dummies(data,drop_first=True).shape
```

# we see that from 6 initial categorial variables , we wnd up with 117 new variables

```python
#  what can we do instead ?
# we can limit one hot encoding to the 10 most frequent labels of the variable.This means that we make one 
# one binary variable for each of the 10 most frequent labels only.
# This is equivalent to grouping all the other labels under a new category,that in this case will be droped.
# thus the 10 new dummy variables indicate if one of the most frequent labels is present (1) or not(0) for a particular observation.

data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# lets make a list woth the most frequent categories of the variable

top_10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
# and now we make the 10 binary variables

for label in top_10:
    data[label] = np.where(data['X2']==label,1,0)
    
data[['X2']+top_10].head(40)   
```

```python
# get whole set of dummy variables , for all the categorial variables

def one_hot_top_x(df , variable , top_x_labels):
    #function to create the dummy variables for the most frequent labels
    #we can vary the number of most frequent that we encode
    
    for label in top_x_labels:
        df[variable+'_'+label] = np.where(data[variable]==label, 1, 0)
    
# read the data again
data = pd.read_csv("C:\\Users\\lenovo\\Downloads\\train.csv.zip" , usecols=['X1','X2','X3','X4','X5','X6'])
    
 # encode X2 into the 10 most frequent categories
one_hot_top_x(data,'X2',top_10)
data.head()
    
```

```python
# advantages 
# Staright forward to implement
# does not require hrs of variable exploration
# Does not expand massively the feature space(number of columns in the dataset)

# Disadvantages
# Does not add any information that may make the variable more predictive
# Does not keep the infomation of the ignored labels.
```
