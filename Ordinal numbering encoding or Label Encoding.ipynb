```python
import pandas as pd
import numpy as np
import datetime
```

```python
df_base = datetime.datetime.today()
```

```python
print(df_base)
```

```python
df_date_list = [df_base - datetime.timedelta(days=x) for x in range(0,20)]
```

```python
print(df_date_list)
```

```python
df = pd.DataFrame(df_date_list)
df.columns=['days']
df
```

```python
df['day_of_week'] = df['days'].dt.day_name()
df.head(10)
```

```python
# categorial variable by ordinal number replacement 
```

```python
weekday_map = {'Monday':1,
            'Tuesday':2,
            'Wednesday':3,
            'Thursday':4,
            'Friday':5,
            'Saturday':6,
            'Sunday':7
               
}
df['day_ordinal'] = df.day_of_week.map(weekday_map)
df.head(20)
```

Statistical analysis is applied to the responses once they are collected to place the people who took the survey into the various categories. The data is then compared to draw inferences and conclusions about the whole surveyed population with regard to the specific variables. The advantages of using ordinal measurement is ease of collation and categorization.If you ask a seuvey question without providing the variables, the answers are likely to be so diverse they cannot be converted to statistics.
